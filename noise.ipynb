```python
import torchshow as ts
gt = "/home/shenss/python/dataset/VOC2012_ORI/train/gt/2007_003565.jpg"
dec = "/home/shenss/python/dataset/VOC2012_ORI/train/evcs/2007_003565.jpg"
cond = "/home/shenss/python/dataset/VOC2012_ORI/train/wsobel/2007_003565.jpg"
ts.save([dec], path="dec.png")
ts.save([gt], path="gt.png")
ts.save([cond],mode='grayscale', path="cond.png")
```

```python
import numpy as np
from PIL import Image

# Load the images
dec = np.array(Image.open(dec))
gt = np.array(Image.open(gt))
res = dec - gt
```

```python
ts.save([res], path='res.png')
noise = np.random.randn(dec.shape[0], dec.shape[1], 3)
ts.save([noise], path='noise.png')
```

```python
a = [0.1, 0.3, 0.5, 0.7, 0.9, 1.0]
for i, x in enumerate(a):
    ts.overlay([dec, noise, gt], alpha=[x, 0.6 * x, 1 - x], save_as=f"fig/{i + 1}.png")
```
