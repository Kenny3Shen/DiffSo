```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import cv2

img = Image.open(img_path)
plt.imshow(img)
```

```python
import torchvision.transforms.v2 as t
# 水平翻转
img_flip = t.RandomHorizontalFlip(p=1)(img)
img_gt_flip = t.RandomHorizontalFlip(p=1)(img)
# 80% jpeg压缩
img_jpeg = t.JPEG(80)(img)

# 添加10%的随机噪声、
noise = np.random.randn(img.size[1], img.size[0], 3) * 0.1 * 255
img_noise = t.Lambda(lambda x: x + noise)(np.array(img))
img_noise = np.clip(img_noise, 0, 255).astype(np.uint8)

#img_gt_flip.save('gt_flip.jpg', format='png')
# img_flip.save('flip.jpg', format='png')
# img_jpeg.save('jpeg.jpg', format='jpeg')
Image.fromarray(img_noise).save('noise.jpg', format='jpeg')




```
